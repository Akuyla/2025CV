```python
from ultralytics import YOLO

# 加载模型
model = YOLO("/kaggle/input/yolov8/pytorch/default/1/yolov8n.pt")

# 训练配置（关键保存参数）
model.train(
    data="/kaggle/input/confignew/VisDrone.yaml",
    epochs=100,
    # 以下为新增的保存和可视化参数
    save=True,                  # 保存训练检查点和结果
    save_period=20,              # 每5个epoch保存一次中间模型
    project="/kaggle/working",  # 所有输出保存到Kaggle工作目录
    name="yolov8_visdrone",     # 实验名称
    exist_ok=True,              # 允许覆盖已有目录
    # 日志和图像记录
    plots=True,                 # 保存训练曲线图
    save_json=True,             # 保存评估结果的JSON文件
    save_hybrid=True,           # 保存混合精度训练日志
    # 验证设置
    val=True,                   # 开启验证
    batch=16,                   # 根据GPU调整batch大小
    imgsz=640,                  # 输入图像尺寸
    # 数据增强（可选）
    augment=True,               # 启用默认数据增强
    mosaic=1.0,                 # 使用Mosaic增强
)
```
